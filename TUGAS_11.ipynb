```
import datetime
import random
import string

data_template = {
    'judul buku': 'blablabla',
    'penulis': 'blablabla',
    'tahun terbit': '0',
    'tanggal terakhir pinjam': datetime.datetime(1111,1,11),
    'status': 'tersedia/tidak'
}

data_buku = {}

while True:
  print(f'{'SELAMAT DATANG':^20}')
  print(f'{'DATA BUKU PERPUSTAKAAN':^20}')
  print('-'*20)

  buku = dict.fromkeys(data_template.keys())
  buku['judul buku'] = input('Judul buku: ')
  buku['penulis'] = input('Nama penulis: ')
  buku['tahun terbit'] = int(input('Tahun terbit: '))

  tahun_terakhir_pinjam = int(input('Tahun terakhir pinjam (YYYY): '))
  bulan_terakhir_pinjam = int(input('Bulan terakhir pinjam (1-12): '))
  tanggal_terakhir_pinjam = int(input('Tanggal terakhir pinjam (1-31): '))
  buku['tanggal terakhir pinjam'] = datetime.datetime(tahun_terakhir_pinjam, bulan_terakhir_pinjam, tanggal_terakhir_pinjam)

  buku['status'] = input('Status (tersedia/tidak): ').lower()

  #meminta sistem untuk membuatkan key secara acak
  KEY = ''.join((random.choice(string.ascii_uppercase) for i in range(6)))

  data_buku.update({KEY: buku})

  print(f'{'KEY':<5} {'judul buku':<30} {'penulis':<25} {'tahun terbit':<15} {'tanggal terakhir pinjam':<25} {'status':<10}')
  print('-'*100)

  for key in data_buku:
    JUDUL = data_buku[key]['judul buku']
    PENULIS = data_buku[key]['penulis']
    TERBIT = data_buku[key]['tahun terbit']
    TANGGAL_TERAKHIR_PINJAM = data_buku[key]['tanggal terakhir pinjam'].strftime('%x')
    STATUS = data_buku[key]['status']

    print(f'{key:<5} {JUDUL:<30} {PENULIS:<25} {str(TERBIT):<15} {TANGGAL_TERAKHIR_PINJAM:<25} {STATUS:<10}')
  print('')

  is_done = input('mau tambah data lagi ga (yes/no)? ')
  if is_done == 'no':
    break

print('Program berakhir')
```
