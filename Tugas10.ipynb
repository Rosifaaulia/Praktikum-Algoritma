2. Dictionary

```python
#Buat dictionary keranjang
Keranjang = {
    'Apel': 5000,
    'Pisang':3000,
    'Jeruk': 7000,
    'Anggur': 15000
}

#Tampilkan setiap barang dan harganya
for buah,harga in Keranjang.items():
  print(f'harga {buah}: Rp {harga}')


#Tampilkan harga total belanja dan gunakan perulangan untuk menghitung total harga semua barang
jumlah = 0
for i in Keranjang.values():
  jumlah += i
print(f'total harga belanja: {jumlah}')
```
